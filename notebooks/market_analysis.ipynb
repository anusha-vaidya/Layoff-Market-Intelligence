# Analyzing the intersection of AI-related layoffs and Wage Velocity.


```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the database
con = duckdb.connect('../data/layoff_intelligence.db')

# Query the AI impact view
df = con.execute("SELECT * FROM v_ai_impact_analysis").df()

print("Sample of AI Impact Data:")
print(df.head())
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='news_volume', y='daily_return_pct', hue='ticker')
plt.title('Market Reaction to AI Layoff News Volume')
plt.show()
```
